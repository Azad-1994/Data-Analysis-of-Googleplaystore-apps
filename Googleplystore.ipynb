### 5. Automatic Categorical

```python
#Total no of apps in each category
categories = {}
for name in data_cl['Category'].unique():
    b=0
    for i in data_cl['Category']:
        if (i == name):
            b +=1
    categories[name]=b

print(categories)
```

```python
# Total no of apps in each type.
Types = {}
for name in data_cl['Type'].unique():
    b=0
    for i in data_cl['Type']:
        if (i == name):
            b +=1
    Types[name]=b

print(Types)
```

```python
# Total no of apps in each content rating.
Content_rating = {}
for name in data_cl['Content Rating'].unique():
    b=0
    for i in data_cl['Content Rating']:
        if (i == name):
            b +=1
    Content_rating[name]=b

print(Content_rating)
```

### Q10. Name the apps that are in ART_AND_DESIGN with rating more than 4.5 in ascending order WRT ratings?
